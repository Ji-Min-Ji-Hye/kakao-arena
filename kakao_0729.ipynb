# 장르당 많이 나오는 태그

```python
#노래이름, 가수기준으로 장르(장르는 벡터화하기),노래아이디 추가
from pandas import Series, DataFrame
import numpy as np
df = pd.DataFrame(columns=['song_name', 'artist_name_basket','tags','top_genre'])  #빈 df선언
for (i,j),group in dataset.groupby(['song_name','artist_name_basket']):#노래별 장르
    a=len(group["song_gn_dtl_gnr_basket"].value_counts())
    list_=[]
    for k in range(0,a):
        list_.append(group["song_gn_dtl_gnr_basket"].value_counts().index[k])   #빈도순으로 태그 추가
    df = df.append(pd.DataFrame([[i, j,list_]], columns=['song_name', 'artist_name_basket','top_genre']), ignore_index=True)
    print(i)
    print(group["song_gn_dtl_gnr_basket"].value_counts())
df.to_csv("song_artist_tag.csv", mode='w')
```
