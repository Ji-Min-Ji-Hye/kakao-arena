```python
import pandas as pd
data_df = pd.read_csv("song_artist_genre_tags_id.csv")
df__=data_df
```

```python
import re
# def preprocessing(sentence):
#     sentence=re.sub('[ ]', ' ', sentence)
#     sentence=sentence.strip()
#     return sentence

# df__['content_cleaned'] = df__["song_gn_gnr_basket"].apply(preprocessing)
content = df__["song_gn_gnr_basket"].tolist()
```

CountVectorizer (API Reference)
: 문서목록에서 각 문서의 feature(문장의 특징) 노출수를 가중치로 설정한 BOW 벡터를 만든다.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
#대분류가 31개
# 군집화 할 그룹의 갯수 정의
n_clusters = 7

# CountVectrizer로 토큰화
vectorizer = CountVectorizer()  #내부원리 확인하기
X = vectorizer.fit_transform(content)

# l2 정규화
X = normalize(X)

# k-means 알고리즘 적용
kmeans = KMeans(n_clusters=n_clusters).fit(X)

# trained labels and cluster centers
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# labels에 merge
df__['labels']=labels
```

```python
df2=df__[['song_gn_gnr_basket', 'labels']]
df2
```

```python
df2=df2.drop_duplicates()
df2
```
