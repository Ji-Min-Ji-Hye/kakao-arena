[CountVectorizer]  

- 가장 단순한 특징으로, 텍스트에서 단위별 등장횟수를 카운팅하여 수치벡터화 하는 것  
단위는 정하는대로입니다.

문서 단위, 문장 단위, 단어 단위 가능  
가장 많이 사용되는 것은 단어단위의 카운팅  

- 카운팅 방법  
: 텍스트 카운팅은 먼저 단어 사전 벡터를 만들고, 카운팅할 문장을 확인하며 그 단어 사전의 횟수를 카운팅하는 것  
예를들면 단어사전에 [나는, 너가, 매일, 공부를, 한다, 좋아한다] 이러한 벡터가 존재한다면,  

'나는 매일 공부를 한다'라는 문장을 카운팅한다면,[1,0,1,1,1,0] 이라는 벡터로 특징을 추출할수 있습니다.  


'나는 매일 매일 공부를 한다'라는 문장이라면,[1,0,2,1,1,0]이라는 벡터를 추출합니다.  


어떤 단어가 많이 발생하는 것에 대해서 어떠한 특징을 지닌다는 법칙을 발견하는 머신러닝 알고리즘의 입력값으로 사용할수 있습니다.  
<br>
<br>
[k-means]  
1단계: 클러스터 개수 k값 선택.  

2단계: 임의로 선택한 k개의 클러스터의 중심과 개별 데이터 사이의 거리 계산 

3단계: 클러스터에 속하게된 데이터들의 평균값을 새로운 클러스터의 중심으로 둠.  

4단계: 2-3단계를 알고리즘이 수렴할때까지 반복. 수렴한다는 것은 클러스터의 중심이 더 이상 변화가 없다는 말  
<br>
<br>

[엘보우 기법]  
SSE값을 비교하면서 최적의 클러스터 개수 찾아줌.


```python
#참고: 같은노래라도 앨범명이 다르면 다르게 저장됨. 
import pandas as pd
data_df = pd.read_csv("song_artist_genre_tags_id.csv")
data_df.drop(['Unnamed: 0'], axis='columns', inplace=True)
data_df
```

```python
#리스트안에 문자열
import re
content = data_df["top_tags"]
a=[]
for i in range(0,len(content)):
    parse = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]','', content[i])
    a.insert(i,parse) #누적
```

```python
len(a)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
# 군집화 할 그룹의 갯수 정의
n_clusters = 100
    
# CountVectrizer로 토큰화
vectorizer = CountVectorizer()  #각 내부 분리해서 적용됨. 문자열마다 벡터 추출
X = vectorizer.fit_transform(a)

# # l2 정규화
# X = normalize(X)
```

```python
# # k-means 알고리즘 적용 [엘버우 기법]으로 클러스터 갯수 찾기
# # kmeans = KMeans(n_clusters=n_clusters).fit(X)
# %matplotlib inline
# from matplotlib import pyplot as plt
# def elbow(X):
#     sse = []
#     for i in range(1,100):
#         km = KMeans(n_clusters=i,algorithm='auto', random_state=42)
#         print(i)
#         km.fit(X)
#         sse.append(km.inertia_)

#     plt.plot(range(1,100), sse, marker='o')
#     plt.xlabel('K')
#     plt.ylabel('SSE')
#     plt.show()

# elbow(X) 
```

```python
kmeans = KMeans(n_clusters=n_clusters).fit(X)

# trained labels and cluster centers
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# labels에 merge
data_df['labels']=labels
```
