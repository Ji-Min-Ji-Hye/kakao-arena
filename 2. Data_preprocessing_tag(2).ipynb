```python
import pandas as pd
dataset = pd.read_csv("All.csv")
dataset.drop(['song_gn_dtl_gnr_basket','Unnamed: 0','Unnamed: 0.1'], axis='columns', inplace=True)
dataset
```

```python
#노래이름, 가수기준으로 태그 카운트
from pandas import Series, DataFrame
import numpy as np
df = pd.DataFrame(columns=['song_name', 'artist_name_basket','top_tags'])  #빈 df선언
for (i,j),group in dataset.groupby(['song_name','artist_name_basket']):
    a=len(group["tags"].value_counts())
    list_=[]
    for k in range(0,a):
        list_.append(group["tags"].value_counts().index[k])   #빈도순으로 태그 추가
    df = df.append(pd.DataFrame([[i, j,list_]], columns=['song_name', 'artist_name_basket','top_tags']), ignore_index=True)
    print(i)
    print(group["tags"].value_counts())
df.to_csv("song_artist_tag.csv", mode='w')
```
